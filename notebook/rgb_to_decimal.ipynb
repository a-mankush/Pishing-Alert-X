```python
import colorgram
import pandas as pd

from color_code import GetRGBValues
from url_screenshot import GetScreenshots, PathConfig

df = pd.read_csv("notebook/data/dataset_B_05_2020.csv").head(500)


class Helpers:
    @staticmethod
    def RGBtoDecimal(rgb) -> int:
        return int("%02x%02x%02x" % (rgb.r, rgb.g, rgb.b), 16)


def getColors(n, i):
    '''
    n is num of colors to extract
    i is specified iindex of that num
    '''
    colors = colorgram.extract("Screenshots/url.png", n)
    print(colors)
    return Helpers.RGBtoDecimal(colors[i].rgb)


def get_deci(url, col_num ,save_screenshot=True):
    ss = GetScreenshots()
    try:
        ss.get_screenshots(url, save_screenshot=True)
        decimal_code = getColors(3, col_num)
    except Exception as e:
        print(f"The {url} is not available")
        decimal_code = 0
    return decimal_code


df["color_1"] = df["url"].apply(lambda x: get_deci(x, 0))
# df["color_2"] = df["url"].apply(lambda x: get_deci(x, 1))
# df["color_3"] = df["url"].apply(lambda x: get_deci(x, 2))

# data['color_1'] = data['hash'].apply(lambda x: getColors(x, 3, 0))

```
